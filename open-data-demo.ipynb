```python
# Import necessary packages
import requests
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Get demographic data
demo_data = requests.get('https://data.cityofnewyork.us/api/odata/v4/kku6-nxdu').json()
demo_df = pd.DataFrame.from_dict(demo_data['value'], orient='columns')

# create a gender graph for a zip code
# subset_df = df #all the zip codes
subset_demo_df = demo_df[demo_df['jurisdiction_name'].isin([10039, 11201])]

new_demo_df = pd.DataFrame({'male': subset_demo_df['count_male'].values, 'female': subset_demo_df['count_female'].values}, index=subset_demo_df['jurisdiction_name'])
new_demo_df.plot.bar(rot=0)
plt.show()
```

```python
# Water Consumption is cool
water_data = requests.get('https://data.cityofnewyork.us/api/odata/v4/ia2d-e54m').json()

water_df = pd.DataFrame.from_dict(water_data['value'], orient='columns')

new_water_df = pd.DataFrame({'water consumption': water_df['nyc_consumption_million_gallons_per_day'].values}, index=water_df['year'])
new_water_df.plot()

new_waterpop_df = pd.DataFrame({'population': water_df['new_york_city_population'].values}, index=water_df['year'])
new_waterpop_df.plot()
plt.show()
```
